```
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split 
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support as score

```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
pick_in2 =open('/content/drive/MyDrive/FinalYear_Project/pickles_dataset/swedishReduced.pickle','rb')
data = pickle.load(pick_in2)
pick_in2.close()
```

```
data
```

```
##Seperating features and labels
import random
random.shuffle(data)
features = []
labels = []

for feature,label in data:
  features.append(feature)
  labels.append(label)
```

```
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.15)
```

```
#Training SVM  with GridSearchCV
# defining parameter range 
param_grid = { 'n_neighbors':[3,5,11,19],
              'weights':['uniform','distance'],
              'metric':['euclidean','manhattan']}  
  
gridKNN = GridSearchCV(KNeighborsClassifier(), param_grid, refit = True, verbose = 3,n_jobs=1) 
  
# fitting the model for grid search 
gridKNN.fit(X_train, y_train)
```

```
print("tuned hpyerparameters :(best parameters) ",gridKNN.best_params_)
print("accuracy :",gridKNN.best_score_)
```

```
#Training KNN on Best fitted Parameters



#Create a KNN Classifier
clf = KNeighborsClassifier(metric='euclidean',n_neighbors=11,weights='distance') 

clf.fit(X_train,y_train)


y_pred = clf.predict(X_test)
acc_knn = round(accuracy_score(y_pred, y_test) * 100, 2)
print("MODEL-2: Accuracy of KNN : ", acc_knn)



precision, recall, fscore, support = score(y_test, y_pred)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))

```
