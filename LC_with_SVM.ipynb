```
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pickle
import random
from sklearn.model_selection import train_test_split 
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_fscore_support as score

```

```
pick_in2 =open('/content/drive/MyDrive/FinalYear_Project/pickles_dataset/swedishReduced.pickle','rb')
data = pickle.load(pick_in2)
pick_in2.close()
```

```
data
```

```
##Seperating features and labels
import random
random.shuffle(data)
features = []
labels = []

for feature,label in data:
  features.append(feature)
  labels.append(label)
```

```
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.15)
```

```
#Training SVM  with GridSearchCV
# defining parameter range 
param_grid = {'C': [0.1, 1, 10, 100, 1000],  
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
              'kernel': ['rbf']}  
  
gridSVM = GridSearchCV(SVC(), param_grid, refit = True, verbose = 3) 
  
# fitting the model for grid search 
gridSVM.fit(X_train, y_train)
```

```
print("tuned hpyerparameters :(best parameters) ",gridSVM.best_params_)
print("accuracy :",gridSVM.best_score_)
```

```
#Training SVM on Best fitted Parameters



#Create a svm Classifier
clf = SVC(C=10, kernel='rbf', gamma=0.01,) 

clf.fit(X_train,y_train)


y_pred = clf.predict(X_test)
acc_svm = round(accuracy_score(y_pred, y_test) * 100, 2)
print("MODEL-2: Accuracy of SVM : ", acc_svm)



precision, recall, fscore, support = score(y_test, y_pred)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
print('fscore: {}'.format(fscore))

```
